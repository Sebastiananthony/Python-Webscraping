```python
from bs4 import BeautifulSoup as beautifulsoup
```

```python
beautifulsoup
```

```python
with open ('fakejob.html', 'r') as html_file:
    content = html_file.read()
    soup = beautifulsoup(content, 'lxml')
    job_class = soup.find_all('div', class_ = 'card-content')
    
    for job in job_class:
        job_title = job.h2.text
        job_company = job.h3.text
        job_location = job.p.text
        post_date = job.time.text
        
        
        more_details = job.a['href']
       
        
        
        print(f'Job title: {job_title.strip()}')
        print(f'Company: {job_company.strip()}')
        print(f'Location: {job_location.strip()}')
        print(f'Posted on: {post_date}')
        print(f'More details: {more_details}')
        
        print(' ')
```
