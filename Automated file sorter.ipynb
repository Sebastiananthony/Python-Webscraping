# Automatic File Sorter in File Explorer

```python
import os, shutil
```

```python
path = r"C:/Users/JUNIPERO/Downloads/"
```

```python
files_name = os.listdir(path)
```

```python
folder_names = ['csv files', 'image files', 'text files', 'exe & xlsx files', 'song files', 'video files']


for loop in range(0, 6):
    if not os.path.exists(path + folder_names[loop]):
        print (path + folder_names[loop])
        os.makedirs(path + folder_names[loop])
```

```python
for file in files_name:
    if '.csv' in file and not os.path.exists(path + "csv files/" + file):
        shutil.move (path + file, path + "csv files/" + file)
    elif '.xlsx' in file and not os.path.exists(path + "csv files/" + file):
        shutil.move (path + file, path + "csv files/" + file)
    elif '.jpg' in file and not os.path.exists(path + "image files/" + file):
        shutil.move (path + file, path + "image files/" + file)
    elif '.png' in file and not os.path.exists(path + "image files/" + file):
        shutil.move (path + file, path + "image files/" + file)
    elif '.txt' in file and not os.path.exists(path + "text files/" + file):
        shutil.move (path + file, path + "text files/" + file)
    elif '.pdf' in file and not os.path.exists(path + "text files/" + file):
        shutil.move (path + file, path + "text files/" + file)
    elif '.exe' in file and not os.path.exists(path + "exe files/" + file):
        shutil.move (path + file, path + "exe files/" + file)
    elif '.mp3' in file and not os.path.exists(path + "song files/" + file):
        shutil.move (path + file, path + "song files/" + file)
    elif '.mkv' in file and not os.path.exists(path + "video files/" + file):
        shutil.move (path + file, path + "video files/" + file)
    elif '.mp4' in file and not os.path.exists(path + "video files/" + file):
        shutil.move (path + file, path + "video files/" + file)
    else:
        print ('Some files in these directory were not sorted due to unspecified file format')
```
